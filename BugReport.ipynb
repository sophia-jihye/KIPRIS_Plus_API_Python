```python
import requests
import xmltodict
import pandas as pd
```

```python
api_key_filepath = 'API_KEY.txt'
with open(api_key_filepath, "r") as f:
    api_key = f.readline()
```

```python
num_of_docs = 10
cpc_number = 'G03F'
ending_yyyy = '2019'

url_format = "http://plus.kipris.or.kr/openapi/rest/patUtiModInfoSearchSevice/cpcSearchInfo"\
                "?cpcNumber={cpc_number}&accessKey={api_key}"\
                "&patent=true&utility=true"\
                "&docsCount={num_of_docs}&sortSpec=OPD&descSort=true&docsStart={page_num}" # 공개일자를 기준으로 내림차순 정렬
```

# Bug
* Page 1: Index 0 ~ 9
* Page 2: Page 1 결과의 index 1 ~ 9 & 새로운 Row 1건
* ...
* Page 9: Page 1 결과의 index 8 ~ 9 & 새로운 Row 8건

### 즉, Page N에서 N이 의미하는 바는 조건에 맞는 전체 문서를 인덱싱할 때, 그 인덱스 번호인 것으로 파악됨
* 한 페이지 당 출력되는 문서 개수는 `K`개이고,
* Index는 0부터 시작되는 상황에서
* Page `N`은 Index `N-1`~`N+K-2`에 해당하는 문서를 출력함

```python
page_num = 1
url = url_format.format(cpc_number=cpc_number, api_key=api_key, page_num=page_num, num_of_docs=num_of_docs)
reponse = requests.get(url)
content = reponse.content
dict_type = xmltodict.parse(content)
item_list = dict_type['response']['body']['items']['PatentUtilityInfo']
pd.DataFrame(item_list)[['ApplicationNumber', 'InventionName', 'Abstract', 'Applicant']]
```
